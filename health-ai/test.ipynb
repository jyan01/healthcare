```python
#ollama 패키지 로드
from langchain_ollama import ChatOllama

#모델 객체화 함수
def create_llm(model="qwen2.5:3b", base_url="http://localhost:11434", temperature=0.2):
    """Ollama 모델 객체를 생성하는 함수"""
    llm_lodel = ChatOllama(
        model=model,
        base_url=base_url,
        temperature=temperature,
    )
    return llm_lodel
    
# 스트리밍 질의 함수
def ask_stream(
     question: str,
    llm_lodel,
):
    """질문을 받아 LLM 응답을 스트리밍으로 출력하는 함수"""
    for chunk in llm_lodel.stream(question):
        if chunk.content:
            yield chunk.content
#사용방법
model = create_llm()
for text in ask_stream("LangChain이 무엇인지 초보자에게 쉽게 설명해줘.", model):
    print(text, end="", flush=True)
```
